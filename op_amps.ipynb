### Inverting Op-Amp
#### Schematic
![](resources/op_amps/inverting_opamp.svg?sanitize=true)

#### Gain Equation
${{Vout = -\dfrac{R2}{R1}Vin}}$

#### Analysis
There are 2 important rules to remember when analyzing an op-amp:
 - No current flows into the input terminals
 - The differential input voltage is 0 as the voltage into non-inverting input and the voltage into inverting input are equal.  
 
Because the voltage at the junction of the feedback loop and the inverting input is the same voltage as the input of the non-inverting input (which is ground), we call this junction a *virtual ground*.  The *virtual ground* makes calculating the input resistance easy as it is equal to the input resistor, R1.  


This *virtual ground* concept also makes analysis of the circuit easier.  We can look at Vout as a f(Vin), and using the above rules, derive the following current equations across each resistor:

${i = \dfrac{Vin - 0}{R1}}$

and 

${i = \dfrac{0-Vout}{R2}}$

Simplifying these and setting them equal to each other, we see that 

${\dfrac{Vin}{R1} = \dfrac{-Vout}{R2}}$

which then simplifies to the above Gain Equation.<br/><br/>
 
When analyzing brute force without using the virtual ground concept, you will instead be writing equations in terms of A where A is the very large gain value of the op-amp.  In those cases, you may get an equation that looks like this:

${AVin = -\dfrac{R1}{R2}Vout\left(1 + A\right) - Vout}$

In this case, you can use the fact that 
 
$Vout\left(1 + A\right)$ 
 
is much greater than $Vout$ and drop the rightmost $Vout$ term.</br></br>

By the same token, 
 
${AVin = -\dfrac{R1}{R2}Vout\dfrac{\left(1 + A\right)}{A}}$ 
 
so you can make 
 
${\dfrac{\left(1 + A\right)}{A}} = 1$ 
 
and land again on our gain equation.<br/><br/>

In the plots below, note how Vout is an inverted copy of Vin scaled by the Gain Equation.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

#simulate input waveform
freq = 2                               # hz
amplitude = 5                          # volts
duration = 2                           # seconds
sample_rate = 100                      # samples per second
num_samples = sample_rate * duration
t = np.arange(num_samples)
Vin = amplitude * np.sin(2 * np.pi * freq * t / num_samples)

# resistor values
R1 = 50000
R2 = 100000

#gain equation
Vout = -R1/R2 * Vin

#plot the results
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))

#common plot settings
for ax in axs.flat:
    ax.grid()
    ax.set_autoscaley_on(False)
    ax.set_ylim([-amplitude, amplitude])
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1.0))
    ax.set_xlabel('time(ms)', fontsize = 14.0)

#Vin plot
axs[0].plot(t, Vin)
axs[0].set_ylabel('Vin(V)', fontsize = 14.0)
axs[0].set_title('Input Voltage', size=18)

#Vout plot
axs[1].plot(t, Vout, 'tab:orange')
axs[1].set_ylabel('Vout(V)', fontsize = 14.0)
axs[1].set_title('Output Voltage', size=18);

#added semicolon above to suppress output to plot window below
```

### Non-Inverting Op-Amp
#### Schematic
![](resources/op_amps/non_inverting_opamp.svg?sanitize=true)

#### Gain Equation
${Vout = \left(1+\dfrac{R2}{R1}\right)Vin}$

#### Analysis
Differences from the inverting op-amp configuration:
 - A non-inverting configuration can never have a gain that is less than 0.
 - Typically the input impedance is higher on a non-inverting configuation because the input voltage is connected directly to the non-inverting input so the input impedance of the op-amp is used (which is typically very high).

### Summing Amplifier
#### Schematic
![](resources/op_amps/summing_amp.svg?sanitize=true)

#### Gain Equation
${Vout = -R4\left(\dfrac{V1}{R1} + \dfrac{V2}{R2} + \dfrac{V3}{R3}\right)}$

If R1 = R2 = R3, then this simplifies to:

${Vout = -\dfrac{R4}{R1}\left(V1 + V2 + V3\right)}$

### Differential Amplifier
#### Schematic
![](resources/op_amps/differential_amp.svg?sanitize=true)

#### Gain Equation
${Vout = -V1\left(\dfrac{R4}{R1}\right) + V2\left(\dfrac{R3}{R2 + R3}\right)\left(\dfrac{R4 + R1}{R1}\right)}$

if R1 = R2 and R3 = R4, then this simplifies to:

${Vout = \dfrac{R4}{R1}\left(V2 - V1\right)}$
